# Number System
 - Binary Number system(base 2)       => 0,1
 - Octal Number system(base 8)        => 0 to 7
 - Decimal Number system(base 10)     => 0 to 9
 - HexaDecimal Number system(base 16) => 0 to 9, a to f(a=10, b=11, c=12, d=13, e=14, f=15)

# Binary Number System(Base 2)

```python
bin(35) # Number to binary conversion
```

```python
int(0b100011) # binaary to number
```

```python
bin(25)
```

```python
int(0b11001)
```
