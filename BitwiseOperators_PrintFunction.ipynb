# Bitwise And(&) Or(|)

```python
bin(12)
bin(13)
print('print binary of 12 is', {bin(12)})
print('print binary of 13 is', {bin(13)})

# 1100 & 1101 = 1100 => which is 12
# so 12 & 13 = 12

12 & 13
```

```python
12 | 13
```

# Xor(^)

```python
12 ^ 13
```

```python
print(bin(25))
print(bin(30))
```

```python
25 ^ 30
```

# Left Shift(<<) -  in Left shift we gain bits

```python
12 << 1 ## in Left shift we gain bits
```

```python
12 << 3

# bin(12) = 0b1100
# 12 << 3 -> means we are adding by default 3 zeros at last
# now 12 << 3 = 1100000
# if we calculate this will get 96
```

```python
int(0b1100000)
```

# Right Shift(>>) - In right shift we loss bits

```python
12 >> 3 

# bin(12) = 0b1100
# 12 >> 3 -> means we are losing by default 3 zeros at last
# now 12 >> 3 = 1
# if we calculate this will get 1 (2 power 0 * 1 = 1)
```

```python
25 >> 2
```

```python
name1 = 'Alice' # B > A
name2 = 'Bob'
name1 < name2
```

```python
name3 = 'Abcd'
name4 = 'BCDA'
name4 > name3
```
