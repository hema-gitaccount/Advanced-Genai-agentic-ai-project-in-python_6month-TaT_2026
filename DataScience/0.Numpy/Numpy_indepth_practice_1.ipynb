# reshape

```python
np.arange(1,11).reshape(5,2) # converted into 5 rows and 2 columns
```

```python
 np.arange(1, 21).reshape(10, 2) # arange will print from 1 to 20 1D array -> then it reshaped to 10 rows and 2 columns
```

```python
np.arange(1,21).reshape(5, 2) # the error bcz you have 20 elements 5 * 2 = 10 only , so it cn't reshape
```

```python
np.arange(1,21).reshape(5, 4) # so now 5 * 4 = 20 now its possible to reshape
```

```python
np.arange(27).reshape(3,3,3)  # 3 power 3 = 27
```
